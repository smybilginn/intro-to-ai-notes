#**MAKİNE ÖĞRENMESİ (MACHINE LEARNING)**
Öncelikle insan öğrenme sürecine bakıcak olursak; insan öğrenmesi deneyimler,gözlemler ve etkileşimler yoluyla bilgi edinme ve beceri geliştirme sürecidir.

Makine öğrenme sürecinde yapay zeka algoritmaları, büyük miktarda veriye dayalı olarak desenleri tanımlar, kararlar alır ve sonuçları tahmin eder.
İnsanların öğrenme sürecine yakın bir şekilde, yapay zeka modelide hatalarında öğrenir ve deneyimleriyle daha iyi hale gelir.

##**Makine Öğrenme Kavramı Nedir?**
Makine öğrenmesindeki _öğrenme_ kavramı, modellerin belirli bir görevi gerçekleştirmek için gerekli olan parametrelerin veya ağırlıkların veriye dayalı olarak ayarlanması veya optimize edilmesini ifade eder.

#**Makine Öğrenmesi Terminolojisi**
1. **Model (Model)**

Model, veriden öğrenilen matematiksel yapıdır. Girdi (özellik) ile çıktı (hedef) arasındaki ilişkiyi öğrenir ve bu ilişkiyi gelecekteki verilerde tahmin yapmak için kullanır.

Örnek:
Ev fiyatlarını tahmin eden bir model, oda sayısı, konum, büyüklük gibi verilere bakarak evin fiyatını tahmin eder.

2. **Öznitelik (Feature)**

Veride yer alan bağımsız değişkenler ya da girdilerdir. Modelin öğrenebilmesi için analiz ettiği bilgiler bunlardır.

Örnek:
Ev fiyatı tahmini problemi için öznitelikler:

Oda sayısı

Metrekare

Mahalle

Bina yaşı

3. **Hedef (Target / Label / Output)**

Modelin öğrenmeye çalıştığı bağımlı değişkendir. Yani çıktıdır. Genellikle tahmin edilmek istenen şeydir.

Örnek:
Ev fiyatı tahmin probleminde hedef:
→ Ev fiyatı (₺ cinsinden)

Sınıflandırma problemlerinde hedef örneği:
→ Email spam mi değil mi? (Spam / Not Spam)

4. **Eğitim (Training)**

Modelin, geçmiş veriler (özellikler ve hedefler) ile öğrenme sürecidir. Model, bu eğitim verilerinden örüntüleri yakalamaya çalışır.

Örnek:
Binlerce evin geçmiş verisiyle model eğitilir. Oda sayısı ve metrekare gibi özellikler ile fiyat arasındaki ilişkiyi öğrenir.

5. **Tahmin (Prediction / İnference)**

Model eğitildikten sonra, yeni verilere dayanarak hedefi tahmin etme işlemidir.

Örnek:
Modelin daha önce görmediği bir evin oda sayısı ve metrekare bilgisi verilirse:
→ “Bu ev yaklaşık 3.200.000₺ eder” gibi bir tahmin yapar.

##**Makine Öğrenmesi Yöntemleri**
1. **Gözetimli Öğrenme (Supervised Learning)**
2. **Gözetimsiz Öğrenme(Unsupervised Learning)**
3. **Pekiştirmeli Öğrenme (Reinfocement Learning)**

### **Gözetimli Öğrenme (Supervised Learning)**
Girdi (özellik) ve doğru çıktı (hedef) verilir. Model, bu ilişkiyi öğrenir.Bu tür öğrenme _etiketlenmiş veri_ ile gerçekleştirilir. Yani her girdiye karşılık gelen bir çıktı etiketi bulunur.

Etiketli veriyle öğrenme (girdi + doğru çıktı → öğren → tahmin et)

Adımları:
1. Veri toplama ve hazırlığı
2. Model seçimi ve eğitimi
3. Model değerlendirmesi
4. Model ayarlaması ve hiperparametre optimizasyonu

Gözetimli öğrenme, sınıflandırma ve regresyon gibi çeşitli problemleri çözmek için kullanılır.Örnek uygulamalar arasında hastalık teşhisi, müşteri tercihlerini tahmin etme bulunabilir.

### **Gözetimsiz Öğrenme(Unsupervised Learning)**
Sadece girdiler verilir, çıktı (hedef) yoktur. Model, veri içindeki yapıyı bulur.

Genellikle veri keşfi ve veri ön işleme aşamasında kullanılır.

Etiketlememiş veri üzerinde yapılan bir makine öğrenmesi yaklaşımıdır.Bu yöntemde, veri kümesindeki yapıyı anlamak, gizli desenleri keşfetmek ve veri noktaları arasındaki ilişkileri belirlemek amaçlanır.

Kümeleme, veri noktalarını içsel özelliklerine göre benzer gruplara ayırmayı amaçlayan bir yöntemdir. Örneğin, K-Means ve DBSCAN algoritmaları yaygın olarak kullanılan kümeleme tekniklerindendir.

Boyut azaltma, yüksek boyutlu verilerdeki gereksiz veya fazla bilgileri ortadan kaldırarak veriyi daha az boyutta temsil etmeyi amaçlar. Bu sayede hem görselleştirme kolaylaşır hem de modelleme süreci daha verimli hale gelir.

Yaygın boyut azaltma algoritmaları PCA (Principal Component Analysis), t-SNE (t-Distributed Stochastic Neighbor Embedding).

### **Pekiştirmeli Öğrenme (Reinforcement Learning)**
Yapay zeka modelinin Ajan (model), bir ortamda deneme-yanılma ile öğrenmesidir. Ödül/kayıp üzerinden gelişir.

Davranış dene → ödül/kayıp al → daha iyi davranmayı öğren

Oyun yapay zekası, rootik ve otomatikleştirilmiş kara alma sistemleri gibi alanlarda yaygın olarak kullanılır.

**Pekiştirmeli Öğrenme Türleri**
- Model Tabanlı Öğrenme (Model-Based RL)
- Modelden Bağımsız Öğrenme (Model-Free RL)
- Politika Tabanlı Öğrenme (Policy-Based)
- Değer Tabanlı Öğrenme (Value-Based)
- Actor-Critic Yöntemleri (Hibrit)

# **Derin Öğrenme (Deep Learning)**
## **Derin Öğrenme Nedir?**
Derin öğrenme bilgisayarlara verileri insan beyninden esinlenerek işlemeyi öğreten bir yapay zeka(AI) yöntemidir.
Derin öğrenme modelleriresimler, metinler,sesler ve diğer karmaşık modelleri tanıyabilir. İnsan davranışını ve zekasını kopyalar.

##**Derin Öğrenme Nasıl Çalışır?**
Derin öğrenme algoritmaları insan beyninden esinlenerek modellenen sinir ağlarıdır.
İnsan beyni bir bilgiyi işlemek için birlikte çalışan milyonlarca nöron içerir.Benzer şekilde derdin öğrenme sinir ağları da bilgisayar içinde birlikte çalışan yapay nöron kaymanından oluşur.
Yapay nöronlar veriyi işlemek için matematiksel hesaplamalar kullanan ve düğüm adı verilen yazılım modülleridir.

##**Derin Öğrenme Ağının Bileşenleri**

Derin öğrenme, yapay sinir ağlarını temel alır. Bu ağlar, aşağıdaki temel bileşenlerden oluşur:


### 1. **Girdi Katmanı (Input Layer)**
- Verinin ağa ilk girdiği katmandır.
- Her nöron bir özelliği temsil eder.

Örnek: El yazısı rakam tanıma için 28x28 piksel = 784 nöron


### 2. **Gizli Katmanlar (Hidden Layers)**
- Girdi ile çıktı arasında yer alır.
- Asıl öğrenme burada gerçekleşir.
- Katman sayısı arttıkça ağ “daha derin” olur (deep learning).

Örnek: Görüntüdeki kenarları, şekilleri, desenleri öğrenir.



### 3. **Çıktı Katmanı (Output Layer)**
- Nihai sonucu üretir.
- Sınıflandırmada her sınıf için bir nöron bulunur.

Örnek: 10 rakam sınıfı için 10 çıkış nöronu (0–9)



### 4. **Nöronlar (Neurons)**
- Her katmandaki temel işlem birimidir.
- Girdi alır, ağırlıkla çarpar, toplar, aktivasyon fonksiyonundan geçirir.



### 5. **Ağırlıklar (Weights)**
- Her bağlantıya atanmış katsayılardır.
- Öğrenme sürecinde güncellenir.
- Ağırlıklar, modelin “öğrendikleridir”.



### 6. **Bias (Sapma)**
- Nöronların daha esnek karar vermesini sağlar.
- Öğrenme sırasında ağırlıklarla birlikte güncellenir.



### 7. **Aktivasyon Fonksiyonu (Activation Function)**
- Nöronun çıktısını hesaplar.
- Karar sınırlarını belirler, ağı doğrusal olmaktan kurtarır.

Yaygın aktivasyon fonksiyonları:
- ReLU (Rectified Linear Unit)
- Sigmoid
- Tanh
- Softmax (özellikle çıktı katmanında)



### 8. **İleri Yayılım (Forward Propagation)**
- Veri, girişten çıkışa doğru katman katman ilerler.
- Çıktı üretilir.



### 9. **Kayıp Fonksiyonu (Loss Function)**
- Modelin yaptığı hatayı ölçer.
- Gerçek sonuçla tahmin arasındaki farktır.
 Örnek:
- Sınıflandırma: Cross-Entropy Loss
- Regresyon: Mean Squared Error (MSE)



### 10. **Geri Yayılım (Backpropagation)**
- Hatanın (kayıp) geriye doğru yayılması.
- Ağırlıklar bu bilgiye göre güncellenir.



### 11. **Optimizasyon Algoritması (Optimizer)**
- Ağırlıkları güncelleyen algoritmadır.
- Amaç: Kayıp fonksiyonunu minimize etmek.
 Yaygın optimizasyon algoritmaları:
- Gradient Descent
- Adam (Adaptive Moment Estimation)

##**Derin Öğrenme Yöntemleri**

Derin öğrenmede farklı yapay sinir ağı mimarileri, farklı veri türleri ve problemler için kullanılır. İşte en yaygın derin öğrenme yöntemleri:


### 1. **Yapay Sinir Ağı (ANN - Artificial Neural Network)**  
- En temel derin öğrenme yapısıdır.  
- Katmanlar ve nöronlardan oluşur.  
- Sayısal veriler, sınıflandırma, regresyon gibi temel görevlerde kullanılır.

Kullanım: Finansal tahmin, tıbbi veri analizi


### 2. **Evrişimli Sinir Ağı (CNN - Convolutional Neural Network)**  
- Görüntü işleme için özel olarak geliştirilmiştir.  
- Evrişim (convolution) ve havuzlama (pooling) katmanları içerir.  
- Görsellerdeki uzamsal ilişkileri öğrenmede çok etkilidir.

Kullanım: Görüntü sınıflandırma, yüz tanıma, nesne tespiti


### 3. **Tekrarlayan Sinir Ağı (RNN - Recurrent Neural Network)**  
- Zaman serisi verilerini işler.  
- Girişler arasında sıralı bağlantılar kurar (geçmiş bilgiyi hatırlar).  
- Metin ve ses gibi ardışık veriler için uygundur.

Kullanım: Dil modeli, ses tanıma, zaman serisi tahmini


### 4. **Uzun Kısa Süreli Bellek (LSTM - Long Short-Term Memory)**  
- RNN’in geliştirilmiş versiyonudur.  
- Uzun vadeli bağımlılıkları daha iyi öğrenir.  
- Bilgi “unutma” ve “hatırlama” mekanizmaları içerir.

Kullanım: Makine çevirisi, metin üretimi, konuşma analizi


### 5. **Transformers**  
- Girdi verisindeki tüm öğeler arasında dikkat (attention) mekanizması kurar.  
- Paralel işlemeye uygundur, çok büyük veriyle etkili çalışır.  
- Dil modellemede devrim yaratmıştır.

Kullanım: ChatGPT, BERT, dil çevirisi, metin anlama


### 6. **Otoenkoder (Autoencoder)**  
- Veriyi daha düşük boyuta indirger, sonra yeniden oluşturur.  
- Boyut azaltma, gürültü giderme ve öznitelik öğrenme için kullanılır.

Kullanım: Anomali tespiti, görsel sıkıştırma


### 7. **Üretici Çekişmeli Ağlar (GAN - Generative Adversarial Networks)**  
- İki ağ birbiriyle yarışır: biri sahte veri üretir, diğeri gerçek mi sahte mi ayırt eder.  
- Yaratıcı yapay zeka uygulamalarında kullanılır.

Kullanım: Sahte görsel üretimi, yüz oluşturma, sanat üretimi


##**Derin Öğrenme Süreçleri ve Araçları**

*Araçlar:* Pandas, NumPy, OpenCV, Scikit-learn, TensorFlow Data API


###  4. **Özellik Mühendisliği**
- Gerekli özelliklerin seçimi/türetilmesi

 *Araçlar:* Scikit-learn, Keras Preprocessing, torchvision.transforms

*Araçlar:*
- **TensorFlow / Keras**

### 6. **Model Değerlendirme**
- Test verisiyle doğruluk, hata oranı gibi metriklerle modelin test edilmesi

#**Üretken Yapay Zeka (Generative AI)**
Üretken yapay zeka, metin, görsel, ses ve sentetik verileri gibi çeşitli içerik türlerini oluşturabilen bir AI teknolojisidir.
Mevcut örneklerden öğrenerek benzer ancak tamamen yeni içerikler oluşturur.


### **Nasıl Çalışır?**  
- Büyük veri setleri üzerinde eğitim alır (örneğin milyonlarca metin, resim)  
- Öğrendiği örüntüleri kullanarak özgün içerik üretir  
- Yaygın teknikler arasında **Generative Adversarial Networks (GANs)** ve **Transformer tabanlı modeller (GPT, DALL·E)** bulunur.


### **Kullanım Alanları**
- Metin yazma (makaleler, şiirler, kod)  
- Görsel oluşturma (sanat, tasarım, fotoğraf)  
- Müzik ve ses üretimi  
- Oyun karakteri, diyalog yaratma  
- İlaç keşfi, simülasyonlar gibi bilimsel uygulamalar
- Müşteri hizmetleri ve teknik destek

### **Örnekler**  
- **ChatGPT:** OpenAI'nin geliştirdiği güçlü bir **dil modeli**  . İnsan benzeri sohbet metinleri oluşturur  
- **DALL·E:** OpenAI tarafından geliştirilen bir **üretken yapay zeka modeli**. Yazılı tariflerden özgün resimler üretir  
- **GAN’lar:** Foto-gerçekçi sahte yüzler veya sanat eserleri yaratır
- **Bard:** Google tarafından geliştirilen bir **yapay zeka sohbet asistanı** .İnsanlarla doğal diyalog kurmak için tasarlanmıştır.İnternet bilgileriyle güncel yanıtlar verebilir, yaratıcı içerikler oluşturabilir

###**Prompt Nedir?**
Prompt, bir modelin bir sorunu veya görevi çözmek için bir giriş formatı veya talimatıdır.


Örneğin:
- "Bir köpekle bir kedinin birlikte piknik yaptığı bir sahne çiz."  → (DALL·E için prompt)
- "Merhaba, bana yapay zekayla ilgili kısa bir metin yazar mısın?" → (ChatGPT için prompt)

### **Prompt Türleri**


### 1. **Bilgi Alma Promtu (Informational Prompt)**
Yapay zekadan bilgi istemek için kullanılır.  
**Örnek:**  
> "Kuantum fiziği nedir?"



###  2. **Yaratıcı Yazım Promtu (Creative Prompt)**
Hikaye, şiir, senaryo gibi yaratıcı metinler üretir.  
**Örnek:**  
> "Bir uzay gemisinde geçen kısa bir bilim kurgu hikayesi yaz."



###  3. **Rol Atama Promtu (Role Prompt)**
Modele bir kimlik veya uzmanlık rolü verilir.  
**Örnek:**  
> "Sen bir tarih öğretmenisin, Fransız Devrimi’ni açıkla."



### 4. **Talimat Verme Promtu (Instructional Prompt)**
Adım adım açıklama yapması istenir.  
**Örnek:**  
> "Python’da nasıl fonksiyon yazılır, adım adım açıkla."



###  5. **Soru-Cevap Promtu (Question & Answer Prompt)**
Kısa ve net bilgi almak için kullanılır.  
**Örnek:**  
> "Dünyanın en yüksek dağı nedir?"



###  6. **Kodlama Promtu (Code Prompt)**
Kod yazdırmak, düzeltmek veya açıklatmak için kullanılır.  
**Örnek:**  
> "Python ile Fibonacci dizisi yaz."


###  7. **Çeviri Promtu (Translation Prompt)**
Metni bir dilden başka bir dile çevirir.  
**Örnek:**  
> "‘Merhaba, nasılsın?’ cümlesini İngilizceye çevir."


###  8. **Özetleme Promtu (Summarization Prompt)**
Uzun metinleri kısa ve öz hale getirir.  
**Örnek:**  
> "Aşağıdaki metni 3 cümle ile özetle."



### 9. **Karşılaştırma Promtu (Comparison Prompt)**
İki kavram ya da şeyin farklarını/similaritelerini listeler.  
**Örnek:**  
> "Makine öğrenmesi ile derin öğrenme arasındaki fark nedir?"



### 10. **Listeleme Promtu (List Prompt)**
Yapay zekadan liste şeklinde bilgi istenir.  
**Örnek:**  
> "Yapay zekanın kullanım alanlarını listele."

###**Promp Tabanlı Öğrenme**
NLP modellerini eğitmek  veya ayarlamak için kullanılan bir yöntemdir.
- Bu yöntemde, modele verilen girişlerin başına özel talimatlar eklenir.
- Bu talimatlar, modele ne tür bir çıktı üretmesi gektiğini belirtir.
- Model bu talimatları dikkate alarak istenilen görevi veya çıktıyı üretmeye çalışır.

#**Yapay Zeka Süreçleri**

### 1. **DevOps (Development + Operations)**
- Yazılım geliştirme (Dev) ve sistem operasyonlarını (Ops) birleştiren yaklaşımdır.  
- Otomasyon, sürekli entegrasyon (CI), sürekli teslim (CD) sağlar.  
- Hızlı, güvenilir ve sürdürülebilir yazılım dağıtımı hedeflenir.


### 2. **MLOps (Machine Learning Operations)**
- DevOps’un makine öğrenmesine uygulanmış halidir.  
- Model geliştirme, test, dağıtım ve izleme süreçlerini otomatikleştirir.  
- Modelin üretim ortamına güvenli ve verimli şekilde geçmesini sağlar.


### 3. **DataOps (Data Operations)**
- Verinin toplanması, temizlenmesi, taşınması ve yönetilmesini otomatikleştirir.  
- Veri akışlarını iyileştirerek veri bilimi ekiplerinin daha hızlı çalışmasına olanak tanır.  
- Kaliteli ve güncel veri sağlamaya odaklanır.


###  4. **AIOps (Artificial Intelligence for IT Operations)**
- BT operasyonlarında yapay zeka kullanımıdır.  
- Log analizi, hata tespiti, sistem izlemede otomasyon sağlar.  
- Sistemlerin kendi kendine öğrenmesini ve sorun çözmesini amaçlar.



### 5. **LLMOps (Large Language Model Operations)**
- Büyük dil modellerinin (ChatGPT, BERT, vs.) yönetimini kapsar.  
- Model güncelleme, veri kontrolü, prompt yönetimi gibi görevleri içerir.  
- LLM’lerin güvenli, verimli ve kontrollü şekilde dağıtımını sağlar.

## ⚖️ **Yapay Zeka Standartları**

Yapay zeka standartları, AI sistemlerinin güvenli, etik ve şeffaf bir şekilde geliştirilmesini ve uygulanmasını sağlayan kurallar bütünüdür.


### **Amaçları**
- Güvenli ve etik AI kullanımı sağlamak  
- Şeffaflık ve hesap verebilirlik oluşturmak  
- Sistemler arası birlikte çalışabilirlik sağlamak  
- Ulusal ve uluslararası uyumluluğu artırmak



### **Önemli Kuruluşlar ve Standartlar**

#### 1. **ISO/IEC (Uluslararası Standartlar Örgütü)**
- Teknik ve yönetsel AI standartları geliştirir  
- Örnekler:  
  - `ISO/IEC 22989` – Yapay zeka terminolojisi  
  - `ISO/IEC 23053` – AI sistem yaşam döngüsü



#### 2. **IEEE (Elektrik ve Elektronik Mühendisleri Enstitüsü)**
- Etik ve sosyal etkiler üzerine odaklanır  
- Örnek: `IEEE 7000 Serisi` – Etik AI sistemleri



#### 3. **OECD (Ekonomik Kalkınma ve İşbirliği Örgütü)**
- AI için küresel ilkeler oluşturur  
- İlke örnekleri: Şeffaflık, kapsayıcılık, insan merkezlilik


#### 4. **EU AI Act (Avrupa Yapay Zeka Yasası)**
- Risk tabanlı yaklaşım benimser  
- Yüksek riskli sistemlere daha sıkı düzenleme getirir



#### 5. **NIST (ABD Ulusal Standartlar Enstitüsü)**
- `AI Risk Management Framework (AI RMF)` geliştirir  
- AI sistemleri için güvenlik ve izlenebilirlik rehberi sunar



### Not:
Standartlar sürekli gelişmektedir ve ülkeler arasında farklılık gösterebilir. Ancak temel hedef: **güvenli, adil ve sorumlu yapay zeka** geliştirmektir.
